```python
#Begin to split data for training
# Check num of batches
len(dataset)
```

```python
# Manually split data into partitions
train_size = int(len(dataset)*.7)
val_size = int(len(dataset)*.2)+1
test_size = int(len(dataset)*.1)
```

```python
train_size
```

```python
val_size
```

```python
train_size + val_size + test_size
```

```python
# Define how much data to take from partition using take and skip methods from tensoflow pipeline
# Make sure date is shuffled
train = dataset.take(train_size)
validate = dataset.skip(train_size).take(val_size)
test = dataset.skip(train_size+val_size).take(test_size)
```

```python
# Build model 
# Import Keras Sequential(one data input/one data output) otherwise use Functional
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
```

```python
# Establish Sequential class
model = Sequential()
```

```python
# Add layers

model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())

model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Conv2D(8, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())

model.add(Flatten())

model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile('adam', loss = tf.losses.BinaryCrossentropy(), metrics= ['accuracy'])
```

```python
# View model
model.summary()
```

```python
# Train model
# Create log directory
logdir='logs'
```

```python
#Create callback to save at specific checkpoints or for specific logging
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=logdir)
```

```python
hist = model.fit(train, epochs=20, validation_data=validate, callbacks=[tensorboard_callback])
```

```python
# Plot model loss
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
# Plot accuracy
fig = plt.figure()
plt.plot(hist.history['accuracy'], color='teal', label='accuracy')
plt.plot(hist.history['val_accuracy'], color='orange', label='val_accuracy')
fig.suptitle('Accuracy', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
#Evaluate Model
#Import dependencies
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator(): 
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    re.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
print(f'Precision: {pre.result().numpy()}, Recall: {re.result().numpy()}, Accuracy: {acc.result().numpy()}')
```

```python
def count_params(model):
    non_trainable_params = np.sum([np.prod(v.get_shape().as_list()) for v in model.non_trainable_weights])
    trainable_params = np.sum([np.prod(v.get_shape().as_list()) for v in model.trainable_weights])
    return {'non_trainable_params': non_trainable_params, 'trainable_params': trainable_params}
```

```python
params = count_params(vgg16_model)
assert params['non_trainable_params'] == 0
assert params['trainable_params'] == 138357544
```
